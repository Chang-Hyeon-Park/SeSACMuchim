## Train

```python
from ultralytics import YOLO

# Load a model
model = YOLO('yolov8n.yaml')
model = YOLO('yolov8l.pt')  # load a pretrained model (recommended for training)

# Train the model
results = model.train(data='yolov8l_pothole_10K.yaml',
                      epochs=150,
                      imgsz=640,              #이미지의 사이즈
                      project = "/content/drive/MyDrive/SeSAC_Final_Project_새싹무침 (1)/model/yolov8/runs",           #결과 저장할 디렉토리 위치
                      name = "Final_yolo_8l", #저장될 폴더 이름
                      exist_ok = False,       #해당 폴더에 결과 존재시 덮어쓰기유
                      lrf=0.001)
```

```python
#test확인
results = model.val(data='yolov8l_pothole_10K.yaml')
```
