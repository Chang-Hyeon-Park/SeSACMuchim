```python
# YOLOv8 모델 로드
model = YOLO("yolov8l.yaml")  # build a new model from scratch
model = YOLO("yolov8l.pt")    # load a pretrained model (recommended for training)
```

```python
# 모델 학습
results = model.train(
    data= "/content/Drive/MyDrive/SeSAC_Final_Project_새싹무침/model/yolov8/yolov8_pothole.yaml",  # 위에서 확인한 yaml 경로 사용
    epochs=100,
    imgsz=640,
    batch=16,
    name='yolov8l_custom',
    device='cpu'  # GPU 사용시 0, CPU 사용시 'cpu'
)
```

```python
results = model.val()  # evaluate model performance on the validation set
```

```python
# 샘플 이미지로 예측 테스트
results = model("/content/Drive/MyDrive/SeSAC_Final_Project_새싹무침/data/TEST_dataset/Pothole/3_pothole_total_619/Pothole_Image_Data/1.jpg") # predict on an image
```

```python
from PIL import Image
# 결과 시각화
for r in results:
    im_array = r.plot()
    im = Image.fromarray(im_array[..., ::-1])
    display(im)
```

```python
# ONNX 형식으로 모델 내보내기
success = model.export(format="onnx")  # 다른 플랫폼에서 사용 가능한 형식으로 변환
# export 성공 여부 확인
print(f"Export successful: {success}")
```
