```python
import xml.etree.ElementTree as ET

# XML 파일 경로 및 출력할 TXT 파일 경로 지정
xml_file_path = '/content/drive/MyDrive/SeSAC_Final_Project_새싹무침/data/pothole_dataset/LARXEL_Kaggle_02/annotations/potholes1.xml'
txt_file_path = '/content/drive/MyDrive/SeSAC_Final_Project_새싹무침/data/pothole_dataset/LARXEL_Kaggle_02/txt/potholes1.txt'

# XML 파일 파싱
tree = ET.parse(xml_file_path)
root = tree.getroot()

# 이미지 크기 추출
width = int(root.find('size/width').text)
height = int(root.find('size/height').text)

# 결과를 YOLO 형식으로 텍스트 파일에 저장
with open(txt_file_path, 'w') as txt_file:
    for obj in root.findall('object'):
        # 클래스 ID 지정 (여기서는 0으로 고정)
        class_id = 0

        # 바운딩 박스 좌표 추출
        bndbox = obj.find('bndbox')
        xmin = int(bndbox.find('xmin').text)
        ymin = int(bndbox.find('ymin').text)
        xmax = int(bndbox.find('xmax').text)
        ymax = int(bndbox.find('ymax').text)

        # 중심 좌표와 너비, 높이를 YOLO 형식으로 변환
        x_center = (xmin + xmax) / 2.0 / width
        y_center = (ymin + ymax) / 2.0 / height
        box_width = (xmax - xmin) / width
        box_height = (ymax - ymin) / height

        # YOLO 형식으로 텍스트 파일에 저장
        txt_file.write(f'{class_id} {x_center:.6f} {y_center:.6f} {box_width:.6f} {box_height:.6f}\n')

print(f"XML 파일에서 모든 포트홀 객체를 YOLO 형식으로 변환하여 {txt_file_path}에 저장했습니다.")
```

```python
import os
import xml.etree.ElementTree as ET

# XML 파일들이 있는 디렉토리 경로와 출력할 TXT 파일들이 저장될 디렉토리 경로 지정
xml_dir_path = '/content/drive/MyDrive/SeSAC_Final_Project_새싹무침/data/pothole_dataset/LARXEL_Kaggle_02/annotations'
txt_dir_path = '/content/drive/MyDrive/SeSAC_Final_Project_새싹무침/data/pothole_dataset/LARXEL_Kaggle_02/txt'

# 디렉토리 내 모든 XML 파일을 순회
for xml_file_name in os.listdir(xml_dir_path):
    if xml_file_name.endswith('.xml'):
        # XML 파일 경로와 동일한 이름의 TXT 파일 경로 지정
        xml_file_path = os.path.join(xml_dir_path, xml_file_name)
        txt_file_name = xml_file_name.replace('.xml', '.txt')
        txt_file_path = os.path.join(txt_dir_path, txt_file_name)

        # XML 파일 파싱
        tree = ET.parse(xml_file_path)
        root = tree.getroot()

        # 이미지 크기 추출
        width = int(root.find('size/width').text)
        height = int(root.find('size/height').text)

        # 결과를 YOLO 형식으로 텍스트 파일에 저장
        with open(txt_file_path, 'w') as txt_file:
            for obj in root.findall('object'):
                # 클래스 ID 지정 (여기서는 0으로 고정)
                class_id = 0

                # 바운딩 박스 좌표 추출
                bndbox = obj.find('bndbox')
                xmin = int(bndbox.find('xmin').text)
                ymin = int(bndbox.find('ymin').text)
                xmax = int(bndbox.find('xmax').text)
                ymax = int(bndbox.find('ymax').text)

                # 중심 좌표와 너비, 높이를 YOLO 형식으로 변환
                x_center = (xmin + xmax) / 2.0 / width
                y_center = (ymin + ymax) / 2.0 / height
                box_width = (xmax - xmin) / width
                box_height = (ymax - ymin) / height

                # YOLO 형식으로 텍스트 파일에 저장
                txt_file.write(f'{class_id} {x_center:.6f} {y_center:.6f} {box_width:.6f} {box_height:.6f}\n')

        print(f"{xml_file_name} 파일을 {txt_file_path}로 변환 완료.")
```

```python
from PIL import Image

# 이미지 파일 및 라벨 파일 경로
image_path = "C:/Users/eeooo/Desktop/Project_road/data2/images194_jpg.rf.dfa7dd7db1b8c73072cc21c0a349f618.jpg"
label_path = "C:/Users/eeooo/Desktop/Project_road/data2/images194_jpg.rf.dfa7dd7db1b8c73072cc21c0a349f618.txt"
output_label_path = "C:/Users/eeooo/Desktop/Project_road/data2/converted_yolo_labels2.txt"

# 이미지 로드 및 크기 가져오기
image = Image.open(image_path)
width, height = image.size

# YOLO 형식으로 변환한 결과를 저장할 리스트
yolo_labels = []

# 라벨 파일 로드
with open(label_path, "r") as f:
    lines = f.readlines()

# 각 라벨의 좌표를 YOLO 형식으로 변환
for line in lines:
    values = line.strip().split()
    class_id = int(values[0])
    points = list(map(float, values[1:]))  # 나머지 좌표 값들

    # x, y 좌표 분리
    x_coords = points[0::2]
    y_coords = points[1::2]

    # 최소, 최대 좌표를 계산하여 바운딩 박스 얻기
    xmin = min(x_coords) * width
    xmax = max(x_coords) * width
    ymin = min(y_coords) * height
    ymax = max(y_coords) * height

    # YOLO 형식 좌표 계산 (정규화)
    x_center = ((xmin + xmax) / 2) / width
    y_center = ((ymin + ymax) / 2) / height
    box_width = (xmax - xmin) / width
    box_height = (ymax - ymin) / height

    # YOLO 형식의 한 줄 추가
    yolo_labels.append(f"{class_id} {x_center:.6f} {y_center:.6f} {box_width:.6f} {box_height:.6f}")

# YOLO 형식의 라벨 파일로 저장
with open(output_label_path, "w") as f:
    for label in yolo_labels:
        f.write(label + "\n")

print("YOLO 형식으로 변환된 라벨 파일이 저장되었습니다.")
```
