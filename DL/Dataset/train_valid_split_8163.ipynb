```python
import os
import random
import shutil

# 1. 데이터 경로 설정
image_dir = "/content/drive/MyDrive/SeSAC_Final_Project_새싹무침/data/Merged_pothole_dataset/images"  # 이미지 파일들이 있는 디렉토리
label_dir = "/content/drive/MyDrive/SeSAC_Final_Project_새싹무침/data/Merged_pothole_dataset/labels"  # 레이블 파일들이 있는 디렉토리
train_image_dir = "/content/drive/MyDrive/SeSAC_Final_Project_새싹무침/data/Merged_pothole_split_dataset/train/images"  # train 이미지 저장 디렉토리
train_label_dir = "/content/drive/MyDrive/SeSAC_Final_Project_새싹무침/data/Merged_pothole_split_dataset/train/labels"  # train 레이블 저장 디렉토리
valid_image_dir = "/content/drive/MyDrive/SeSAC_Final_Project_새싹무침/data/Merged_pothole_split_dataset/valid/images"  # valid 이미지 저장 디렉토리
valid_label_dir = "/content/drive/MyDrive/SeSAC_Final_Project_새싹무침/data/Merged_pothole_split_dataset/valid/labels"  # valid 레이블 저장 디렉토리

# 2. 이미지와 레이블 파일 매칭
image_files = sorted([f for f in os.listdir(image_dir) if os.path.isfile(os.path.join(image_dir, f))])
label_files = sorted([f for f in os.listdir(label_dir) if os.path.isfile(os.path.join(label_dir, f))])

# 매칭 확인 (파일 이름이 같아야 함)
assert len(image_files) == len(label_files), "Image and label counts do not match!"
for img, lbl in zip(image_files, label_files):
    assert os.path.splitext(img)[0] == os.path.splitext(lbl)[0], f"File mismatch: {img} and {lbl}"

# 3. 데이터 병합 및 셔플
data_pairs = list(zip(image_files, label_files))
random.shuffle(data_pairs)  # 이미지와 레이블 파일을 함께 셔플

# 4. Train/Validation 데이터셋 비율 계산
train_size = int(len(data_pairs) * 0.8)  # 80%는 train
train_pairs = data_pairs[:train_size]
valid_pairs = data_pairs[train_size:]

# 5. 저장 디렉토리 생성
os.makedirs(train_image_dir, exist_ok=True)
os.makedirs(train_label_dir, exist_ok=True)
os.makedirs(valid_image_dir, exist_ok=True)
os.makedirs(valid_label_dir, exist_ok=True)

# 6. 파일 복사
for img_file, lbl_file in train_pairs:
    shutil.copy(os.path.join(image_dir, img_file), os.path.join(train_image_dir, img_file))
    shutil.copy(os.path.join(label_dir, lbl_file), os.path.join(train_label_dir, lbl_file))

for img_file, lbl_file in valid_pairs:
    shutil.copy(os.path.join(image_dir, img_file), os.path.join(valid_image_dir, img_file))
    shutil.copy(os.path.join(label_dir, lbl_file), os.path.join(valid_label_dir, lbl_file))

print(f"Total {len(train_pairs)} pairs in train dataset.")
print(f"Total {len(valid_pairs)} pairs in validation dataset.")
```
