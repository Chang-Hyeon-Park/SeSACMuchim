```python
# 서현님 여기에요~
import os
import shutil

# 이미지와 라벨 파일을 순회하며 번호를 부여하여 새 이름으로 복사
def convert_file_name(image_dir, label_dir, file_count):
    for image_file in sorted(os.listdir(image_dir)):
        # 이미지 파일 확장자 확인
        if image_file.endswith(".jpg") or image_file.endswith(".png"):
            # 기존 이미지와 라벨 파일 경로
            image_path = os.path.join(image_dir, image_file)
            label_path = os.path.join(label_dir, os.path.splitext(image_file)[0] + ".txt")

            # 새 파일 이름 설정 (기존 이미지 확장자 유지)
            new_image_name = f"pothole_{file_count}{os.path.splitext(image_file)[1]}"
            new_label_name = f"pothole_{file_count}.txt"

            # 새로운 경로 생성
            new_image_path = os.path.join(new_image_dir, new_image_name)
            new_label_path = os.path.join(new_label_dir, new_label_name)

            # 파일이 존재하는 경우에만 복사 수행
            if os.path.exists(label_path):
                # 파일 복사 및 이름 변경
                shutil.copy(image_path, new_image_path)
                shutil.copy(label_path, new_label_path)
                print(f"{image_file} -> {new_image_name}, {os.path.basename(label_path)} -> {new_label_name}")
                file_count += 1
            else:
                print(f"{label_path} 라벨 파일이 존재하지 않습니다.")
```

```python
import os
import random
import shutil



def split_and_match_dataset(image_folder, txt_folder, output_folder, train_ratio=0.7, valid_ratio=0.2, test_ratio=0.1):
    """
    이미지 폴더와 텍스트 파일 폴더를 7:2:1 비율로 섞어서 train, valid, test로 나눕니다.
    - image_folder: 이미지 파일들이 있는 폴더
    - txt_folder: 텍스트 파일들이 있는 폴더
    - output_folder: 결과를 저장할 폴더
    파일 move!!!!
    """
    # 이미지와 라벨 파일의 공통 이름 찾기
    image_files = {os.path.splitext(f)[0]: f for f in os.listdir(image_folder) if f.lower().endswith(('png', 'jpg', 'jpeg'))}
    txt_files = {os.path.splitext(f)[0]: f for f in os.listdir(txt_folder) if f.lower().endswith('.txt')}

    # 공통된 파일 이름 매칭
    common_files = list(image_files.keys() & txt_files.keys())
    random.shuffle(common_files)  # 섞기

    # 전체 개수 계산
    total_count = len(common_files)
    train_count = int(total_count * train_ratio)
    valid_count = int(total_count * valid_ratio)
    test_count = total_count - train_count - valid_count

    # 파일 분할
    train_files = common_files[:train_count]
    valid_files = common_files[train_count:train_count + valid_count]
    test_files = common_files[train_count + valid_count:]

    # 폴더 생성
    train_image_folder = os.path.join(output_folder, 'train/images')
    valid_image_folder = os.path.join(output_folder, 'valid/images')
    test_image_folder = os.path.join(output_folder, 'test/images')
    train_txt_folder = os.path.join(output_folder, 'train/labels')
    valid_txt_folder = os.path.join(output_folder, 'valid/labels')
    test_txt_folder = os.path.join(output_folder, 'test/labels')

    os.makedirs(train_image_folder, exist_ok=True)
    os.makedirs(valid_image_folder, exist_ok=True)
    os.makedirs(test_image_folder, exist_ok=True)
    os.makedirs(train_txt_folder, exist_ok=True)
    os.makedirs(valid_txt_folder, exist_ok=True)
    os.makedirs(test_txt_folder, exist_ok=True)

    # 파일 이동 함수
    def move_files(file_list, image_dest, txt_dest):
        for file in file_list:
            # 이미지 이동
            src_image_path = os.path.join(image_folder, image_files[file])
            dst_image_path = os.path.join(image_dest, image_files[file])
            shutil.move(src_image_path, dst_image_path)
            print(f"[IMAGE MOVED] {src_image_path} -> {dst_image_path}")

            # 라벨 이동
            src_txt_path = os.path.join(txt_folder, txt_files[file])
            dst_txt_path = os.path.join(txt_dest, txt_files[file])
            shutil.move(src_txt_path, dst_txt_path)
            print(f"[LABEL MOVED] {src_txt_path} -> {dst_txt_path}")

    # 파일 이동
    print("[STARTING TRAIN FILES MOVE]")
    move_files(train_files, train_image_folder, train_txt_folder)
    print("[STARTING VALID FILES MOVE]")
    move_files(valid_files, valid_image_folder, valid_txt_folder)
    print("[STARTING TEST FILES MOVE]")
    move_files(test_files, test_image_folder, test_txt_folder)

    print(f"Dataset split completed! Total: {total_count}, Train: {train_count}, Valid: {valid_count}, Test: {test_count}")

# 경로 설정
image_folder = '/content/drive/MyDrive/SeSAC_Final_Project_새싹무침/data/Ai_hub_Busan_dataset/Busan_Merged_dataset_rename_images'  # 이미지 파일이 있는 폴더
txt_folder = '/content/drive/MyDrive/SeSAC_Final_Project_새싹무침/data/Ai_hub_Busan_dataset/Busan_Merged_dataset_rename_labels'  # 텍스트 파일이 있는 폴더
output_folder = '/content/drive/MyDrive/SeSAC_Final_Project_새싹무침/data/Ai_hub_Busan_dataset/Busan_Merged_dataset_split'  # 결과를 저장할 폴더

# 함수 호출
split_and_match_dataset(image_folder, txt_folder, output_folder)
```
