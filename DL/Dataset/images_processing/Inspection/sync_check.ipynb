```python
import os

# 특정 디렉토리 경로 설정 (현재 디렉토리를 기준으로 할 경우 '.' 사용)
directory_path = '/content/drive/MyDrive/SeSAC_Final_Project_새싹무침/data/pothole_dataset/AIHub_도로장애물_포트홀(35,651)42GB/sample5000/annotations'

# '(1)'이 포함된 파일명을 저장할 리스트 초기화
file_list_with_1 = []

# 디렉토리 내 파일들을 순회하며 '(1)'이 포함된 파일명을 리스트에 추가
for filename in os.listdir(directory_path):
    if '(1)' in filename:
        file_list_with_1.append(filename)

# 파일 리스트에 있는 파일들을 삭제
for filename in file_list_with_1:
    file_path = os.path.join(directory_path, filename)
    try:
        os.remove(file_path)
        print(f"{filename} 파일이 삭제되었습니다.")
    except FileNotFoundError:
        print(f"{filename} 파일을 찾을 수 없습니다.")
    except Exception as e:
        print(f"{filename} 파일 삭제 중 오류 발생: {e}")
```
