```python
from PIL import Image

# 이미지 파일 및 라벨 파일 경로
image_path = "/content/drive/MyDrive/SeSAC_Final_Project_새싹무침/data/pothole_dataset/VITVEL_ROBO_01/test/images"
label_path = "/content/drive/MyDrive/SeSAC_Final_Project_새싹무침/data/pothole_dataset/VITVEL_ROBO_01/test/labels"
output_label_path = "/content/drive/MyDrive/SeSAC_Final_Project_새싹무침/data/pothole_dataset/VITVEL_ROBO_01/test/txt"

# 이미지 로드 및 크기 가져오기
image = Image.open(image_path)
width, height = image.size

# YOLO 형식으로 변환한 결과를 저장할 리스트
yolo_labels = []

# 라벨 파일 로드
with open(label_path, "r") as f:
    lines = f.readlines()

# 각 라벨의 좌표를 YOLO 형식으로 변환
for line in lines:
    values = line.strip().split()
    class_id = int(values[0])
    points = list(map(float, values[1:]))  # 나머지 좌표 값들

    # x, y 좌표 분리
    x_coords = points[0::2]
    y_coords = points[1::2]

    # 최소, 최대 좌표를 계산하여 바운딩 박스 얻기
    xmin = min(x_coords) * width
    xmax = max(x_coords) * width
    ymin = min(y_coords) * height
    ymax = max(y_coords) * height

    # YOLO 형식 좌표 계산 (정규화)
    x_center = ((xmin + xmax) / 2) / width
    y_center = ((ymin + ymax) / 2) / height
    box_width = (xmax - xmin) / width
    box_height = (ymax - ymin) / height

    # YOLO 형식의 한 줄 추가
    yolo_labels.append(f"{class_id} {x_center:.6f} {y_center:.6f} {box_width:.6f} {box_height:.6f}")

# YOLO 형식의 라벨 파일로 저장
with open(output_label_path, "w") as f:
    for label in yolo_labels:
        f.write(label + "\n")

print("YOLO 형식으로 변환된 라벨 파일이 저장되었습니다.")
```
