```python
# Test Data Merge
import os
import random
import shutil

# 1. 데이터 디렉토리 경로 설정
plain_dir = "/content/drive/MyDrive/SeSAC_Final_Project_새싹무침/data/TEST_dataset/Plain/18_plain_total_352"  # plain 디렉토리 경로
pothole_dir_1 = "/content/drive/MyDrive/SeSAC_Final_Project_새싹무침/data/TEST_dataset/Pothole/18_pothole_total_696"  # pothole 첫 번째 디렉토리
pothole_dir_2 = "/content/drive/MyDrive/SeSAC_Final_Project_새싹무침/data/TEST_dataset/Pothole/3_pothole_total_619"  # pothole 두 번째 디렉토리
output_dir = "/content/drive/MyDrive/SeSAC_Final_Project_새싹무침/data/Merged_pothole_test_dataset"  # 병합된 데이터 저장 경로

# 2. 파일 목록 불러오기
plain_files = [os.path.join(plain_dir, f) for f in os.listdir(plain_dir) if os.path.isfile(os.path.join(plain_dir, f))]
pothole_files_1 = [os.path.join(pothole_dir_1, f) for f in os.listdir(pothole_dir_1) if os.path.isfile(os.path.join(pothole_dir_1, f))]
pothole_files_2 = [os.path.join(pothole_dir_2, f) for f in os.listdir(pothole_dir_2) if os.path.isfile(os.path.join(pothole_dir_2, f))]

# 3. 병합 및 셔플
merged_files = plain_files + pothole_files_1 + pothole_files_2
random.shuffle(merged_files)

# 4. 저장 디렉토리 생성
os.makedirs(output_dir, exist_ok=True)

# 5. 파일 네이밍 및 복사
for idx, file_path in enumerate(merged_files, start=1):
    # 새로운 파일 이름 설정
    new_filename = f"pothole_test_{idx}{os.path.splitext(file_path)[1]}"
    new_file_path = os.path.join(output_dir, new_filename)

    # 파일 복사
    shutil.copy(file_path, new_file_path)

print(f"Total {len(merged_files)} files merged and saved in {output_dir}.")
```
