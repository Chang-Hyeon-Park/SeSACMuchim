```python
import os
'''
7937 -> 4782
7936 -> 7493
'''


def rename_file(directory, old_name, new_name):
    # old_name과 new_name에 접두사 'pothole_'를 추가하여 파일 이름을 구성
    old_filename = f"pothole_{old_name}"
    new_filename = f"pothole_{new_name}"

    # 디렉토리 내 파일 목록을 순회하며 파일 이름 변경
    for filename in os.listdir(directory):
        # 확장자 포함하여 파일 이름을 비교
        if filename.startswith(old_filename):
            # 파일의 확장자 추출
            ext = os.path.splitext(filename)[1]

            # 기존 파일 경로와 새 파일 경로 정의
            old_path = os.path.join(directory, filename)
            new_path = os.path.join(directory, f"{new_filename}{ext}")

            # 파일 이름 변경
            os.rename(old_path, new_path)
            print(f"'{old_filename}{ext}'를 '{new_filename}{ext}'로 변경했습니다.")
            return

    print(f"'{old_filename}' 파일을 찾을 수 없습니다.")

# 예시 사용법
directory_path = '/content/drive/MyDrive/SeSAC_Final_Project_새싹무침/data/Merged_pothole_dataset/labels'  # 파일이 있는 디렉토리 경로
rename_file(directory_path, 7937, 4782)
```
