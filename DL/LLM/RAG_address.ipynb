```python
from langchain.embeddings import SentenceTransformerEmbeddings

embeddings = SentenceTransformerEmbeddings(model_name="all-MiniLM-L6-v2")
```

```python
from langchain.docstore.document import Document
from langchain.vectorstores import Chroma

# 예시 데이터 (JSON 파일에서 불러온 relevant_data)
documents = [Document(page_content="강남구 남부순환로 17길에서 발견", metadata={"file_name": "pothole_78"})]
```

```python
vectorstore = Chroma.from_documents([documents], embedding=embeddings)
```

```python
from langchain.llms import OpenAI

# OpenAI API 키 설정
import os
os.environ["OPENAI_API_KEY"] = 

# LLM 연결
llm = OpenAI(temperature=0.7)

# 질문
query = "pothole_100 이 발견된 곳은 어디야?"

# 벡터 스토어에서 유사한 문서 검색
docs = vectorstore.similarity_search(query, k=1)

# LLM에 질문과 검색된 내용을 전달하여 응답 생성
context = docs[0].page_content
response = llm(f"질문: {query}\n문맥: {context}\n답변:")
print(response)
```

```python
places = []
for i in range(len(relevant_data)):
    documents = Document(page_content= f"f{relevant_data.loc[i]['구']}  {relevant_data.loc[i]['로']} 에서 {relevant_data.loc[i]['file_name']}번 포트홀 발견",
                         metadata={"file_name":relevant_data.loc[i]['file_name'] })
    places.append(documents)
```

```python
len(places)
```

```python
vectorstore = Chroma.from_documents(places, embedding=embeddings)
```

```python
# 질문
query = "pothole_100의 장소는?"

# 벡터 스토어에서 유사한 문서 검색
docs = vectorstore.similarity_search(query, k=1)

# LLM에 질문과 검색된 내용을 전달하여 응답 생성
context = docs[0].page_content
response = llm(f"질문: {query}\n문맥: {context}\n답변:")
print(response)
```
