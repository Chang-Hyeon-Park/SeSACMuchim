```python
# 3번의 질문을 입력받아 처리
for i in range(3):  # 3번 반복
    user_query = input(f"질문 {i+1}/3을 입력하세요: ")  # 질문 입력받기
    if not user_query.strip():  # 입력이 없을 경우 처리
        print("질문을 입력하지 않았습니다. 다음 질문으로 넘어갑니다.")
        continue
    # Query 실행
    response = query_engine.query(user_query)  # temperature 제거
    # 결과 출력
    print(f"질문 {i+1}에 대한 응답:\n{str(response)}\n")
```
